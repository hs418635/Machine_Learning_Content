```python
##bar chart

import numpy as np
import matplotlib.pyplot as plt
object = ('p','y','t','h','o','n')
y_pos = np.arange(len(object))
y_pos
l = [10,8,6,4,2,1]
plt.ylabel("usage")
plt.title("programming language used")
plt.bar(y_pos,l, align = 'center',alpha=0.9)
plt.show()
```

```python
#horizontal graph
import matplotlib.pyplot as plt
import numpy as np
objects = ('p','y','t','h','o','n')
y_pos = np.arange(len(objects))
l = [10,8,6,4,2,1]
plt.barh(y_pos,l,align ='center',alpha=0.8)
plt.yticks(y_pos,objects)
plt.xlabel('usage')
plt.title("programming language used")
plt.show()
```

```python
#pi chart
labels =  ('p','y','t','h','o','n')
sizes = (215,130,245,210,200,100)
colors = ['gold','red','skyblue','yellow','green','black']
e = (0,0,0,0,0,0.1)
plt.pie(sizes,explode = e, labels= labels,colors=colors,autopct='%.2f',shadow=True)
#%matplotlib auto
#plt.pie(sizes,labels=labels,autopct="%.2f")
plt.axis('equal')
plt.show()
plt.tight_layout()
plt.draw()
```
