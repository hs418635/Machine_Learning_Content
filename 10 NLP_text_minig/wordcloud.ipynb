```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
reader_ob = open(r"E:\DebugShala\Machine learning\10 NLP_text_minig\text_file.txt")
reader_content = list(reader_ob)
```

```python
reader_content
```

```python
text = ""

for row in reader_content:
    for word in row:
        text = text + "" + word
wordcloud = WordCloud(width=580, height = 580 , max_words=100).generate(text)
```

```python
# ploting the wordcloud image
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("on")
plt.margins(x=10 , y=0)
```
