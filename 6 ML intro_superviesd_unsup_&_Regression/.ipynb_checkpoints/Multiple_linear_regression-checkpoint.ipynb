```python
## reshuffling the columns
data1 = data.reindex(columns=['Country', 'Year', 'Status', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income composition of resources', 'Schooling','Life expectancy '])
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1.isna().sum()
```

```python
df = data1.dropna()
```

## dealing with categorical data

```python
df.columns
```

```python
cat_col = ['Country',"Status"]
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
for i in cat_col:
    df[i] = lb.fit_transform(df[i])
df[i]
```

```python
df.head()
```

## dealing with cont_columns

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
```

```python
con_col = ['Year', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income composition of resources', 'Schooling']
```

```python
for i in con_col:
    df[i] = mm.fit_transform(df[i].values.reshape(-1,1))
df[i].head()
```

```python
df.head()
```

```python
### splitting into train test
from sklearn.model_selection import train_test_split
train, test = train_test_split(df,test_size=0.2)
```

```python
df.shape
```

```python
train.shape
```

```python
test.shape
```

```python
### separating independent and dependent variables

train_x = train.iloc[:,:-1] # independent
train_y = train.iloc[:,-1] # dependent
```

```python
test_x = test.iloc[:,:-1] # independent
test_y = test.iloc[:,-1] # dependent
```

```python
## simple linear regression

from sklearn.linear_model import LinearRegression
mlr = LinearRegression().fit(train_x,train_y)
```

```python
prd = mlr.predict(test_x)
prd
```

```python
test_y
```

```python
# rmse

import numpy as np

rmse = np.sqrt(np.mean(prd-test_y)**2)
```
