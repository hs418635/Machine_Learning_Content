```python
# shuffling the target column into last place
df1 = df.reindex(columns=['Country', 'Year', 'Status', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income composition of resources', 'Schooling','Life expectancy '])
df1.head()
```

```python
# dealing with cat_columns
cat_col = ["Country","Status"]

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for i in cat_col:
    df1[i] = lb.fit_transform(df1[i])
```

```python
# dealing with continuous columns
con_col = ['Year', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income composition of resources', 'Schooling']
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for i in con_col:
    df1[i] = mm.fit_transform(df[i].values.reshape(-1,1))
```

### Filling Null Values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
imputer.fit(df1)
```

```python
df1_imputed = pd.DataFrame(imputer.transform(df1), columns=df1.columns)
df1_imputed.head()
```
